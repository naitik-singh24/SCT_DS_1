```python
import matplotlib.pyplot as plt

# Sample data
genders = ['Male', 'Female', 'Other']
counts = [120, 100, 15]

# Create bar chart
plt.figure(figsize=(6, 4))
plt.bar(genders, counts)

plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of People')

plt.savefig('gender_distribution.png') #Save as PNG

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample age data
ages = [18, 19, 20, 21, 22, 22, 23, 24, 25, 25,
        26, 27, 28, 29, 30, 31, 32, 35, 40, 45]

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(ages, bins=5)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.savefig('age_distribution.png') #Save as PNG

plt.show()

```

```python
# Both BAR CHART And HISTOGRAM Are Presented

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
file_name = list(uploaded.keys())[0]

df = pd.read_excel(
    file_name,
    sheet_name="Data",
    skiprows=3
)

# Find the latest available year
year = df.columns[-1]

# Keep only Country Name and Population
population = df[['Country Name', year]].dropna()

# ---------------------------
# Bar Chart (Top 10 Countries)
# ---------------------------
top10 = population.sort_values(by=year, ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top10['Country Name'], top10[year], color='skyblue')
plt.title(f"Top 10 Most Populated Countries ({year})")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# ---------------------------
# Histogram
# ---------------------------
plt.figure(figsize=(10,6))
plt.hist(population[year], bins=30, color='orange', edgecolor='black')
plt.title(f"Population Distribution ({year})")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
